# Bienvenidos!

**Bienvenidos en el OGGM-Edu JupyterLab**! Aquí se puede realizar experimentos con "glaciares digitales": sin hielo pero interactivos y menos peligrosos que los glaciares Peruanos.

Es probable que ese entorno "JupyterLab" es nuevo para ti, pero no te preocupes! Aquí nunca te puedes equivocar: siempre se puede empezar de nuevo.

En esta pagina te enseñamos las características fundamentales del entorno "JupyterLab" de OGGM-Edu: para mas explicaciones, escucha las recomendaciones de tu profesora.

## Estructura

El JupyterLab tiene tres zonas principales: la **barra del menú** (azul), el **explorador de archivos** (rojo), y las **pestañas con los cuadernos activos** (verde):

<img src="http://edu.oggm.org/en/latest/_images/docs_binder_jlab.png" align="left" width="70%">

## Abrir, cerrar y guardar los cuadernos

En el **explorador de archivos** se puede seleccionar los cuadernos: haz un "doble clic" en el cuaderno que quieres abrir.

Para cerrar un cuaderno, usa la cruz de la pestaña (como en tu navegador web). Si haz cambiado algo en tu cuaderno, veras que la cruz esta escondida por un circulo gris. Para guardar tus cambios, usa el botón como enseñado en la video abajo, o usa `CTRL+S`.

## Copiar un archivo

Copiar los cuadernos antes de cambiarles puede ser muy útil: por ejemplo para hacer experimentos sin perder los archivos originales.

**Por eso te recomendamos de renombrar los cuadernos antes de hacer los experimentos**. Así, siempre puedes volver a ver al cuaderno original. Copiar un cuaderno es muy fácil: haz un clic derecho en el archivo y selecciona "Duplicate". Despues, renombra el archivo como te lo enseña esta video: 

*Duplicate/rename/edit/delete video*:

<video controls src="img/duplicate.mp4" />

## Volver al estado original

Si por algún razón quieres volver al estado original del curso (por ejemplo si has perdido los cuadernos originales, o si la profesora te recomendó de hacerlo), es muy simple:

1. Borra o renombra el directorio de trabajo
2. Cerra tu entorno "JupyterLab"
3. Ábrelo otra vez desde la pagina web CdeC-glaciologia: https://github.com/ehultee/CdeC-glaciologia
